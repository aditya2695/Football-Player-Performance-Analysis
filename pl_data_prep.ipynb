```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
attack_df =  pd.read_excel('data/premier_league_complete.xlsx',sheet_name='attack')
defence_df =  pd.read_excel('data/premier_league_complete.xlsx',sheet_name='defence')
passing_df =  pd.read_excel('data/premier_league_complete.xlsx',sheet_name='passing')
possession_df =  pd.read_excel('data/premier_league_complete.xlsx',sheet_name='possession')
shooting_df =  pd.read_excel('data/premier_league_complete.xlsx',sheet_name='shooting')
standard_df = pd.read_excel('data/premier_league_complete.xlsx',sheet_name='standard')
```

```python
standard_df.drop(columns=['Matches'],inplace=True)
shooting_df.drop(columns=['Player','PK','PKatt','xG','npxG','90s','Gls','Matches','Nation','Pos','Squad','Age','Born'],inplace=True)
attack_df.drop(columns=['Player','90s','Sh','Matches','Nation','Pos','Squad','Age','Born'],inplace=True)
defence_df.drop(columns=['Player','90s','Sh','Matches','Nation','Pos','Squad','Age','Born'],inplace=True)
passing_df.drop(columns=['Player','Att','Ast','xAG','PrgP','90s','Matches','Nation','Pos','Squad','Age','Born'],inplace=True)
possession_df.drop(columns=['Player','90s','TotDist','final_third','PrgDist','PrgC','Att 3rd','Mid 3rd','Att','Def 3rd','PrgR','Matches','Nation','Pos','Squad','Age','Born'],inplace=True)

```

```python
print('standard_df shape:',standard_df.shape)
print('shooting shape:',shooting_df.shape)
print('attack_df shape:',attack_df.shape)
print('defence_df shape:',defence_df.shape)
print('passing_df shape:',passing_df.shape)
print('possession_df shape:',possession_df.shape)
```

```python
shooting_df.drop_duplicates(inplace=True)
attack_df .drop_duplicates(inplace=True)
defence_df .drop_duplicates(inplace=True)
print('standard_df shape:',standard_df.shape)
print('shooting shape:',shooting_df.shape)
print('attack_df shape:',attack_df.shape)
print('defence_df shape:',defence_df.shape)
print('passing_df shape:',passing_df.shape)
print('possession_df shape:',possession_df.shape)
```

```python
shooting_df  = shooting_df.drop(index=25)
attack_df  = attack_df .drop(index=25) 
defence_df  = defence_df .drop(index=25)

print('standard_df shape:',standard_df.shape)
print('shooting shape:',shooting_df.shape)
print('attack_df shape:',attack_df.shape)
print('defence_df shape:',defence_df.shape)
print('passing_df shape:',passing_df.shape)
print('possession_df shape:',possession_df.shape)

```

```python
standard_df = standard_df.set_index('Rk')
shooting_df = shooting_df.set_index('Rk')
attack_df = attack_df.set_index('Rk')
defence_df = defence_df.set_index('Rk')
passing_df = passing_df.set_index('Rk')
possession_df = possession_df.set_index('Rk')
```

**Now we can start joining the dataset**

```python
# Horizontally stack the DataFrames
combined_df = pd.concat([standard_df, shooting_df, attack_df, defence_df, passing_df, possession_df], axis=1)
combined_df.shape
```
