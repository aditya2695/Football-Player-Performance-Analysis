## **Get Released Players**

```python
# Read the text file
with open('data/championship_released.txt', 'r') as file:
    content = file.readlines()

# Create a dictionary to store released players and their previous clubs
released_players = {}
current_team = None

# Process each line in the text file
for i in range(0, len(content), 2):
    team = content[i].strip()
    players = content[i + 1].strip().split(':')[1].strip().split(', ')
    released_players[team] = players

# Print the released players and their previous clubs
print(released_players)
```

```python
released_count=0
for team, players in released_players.items():
    for player in players:
        released_count+=1
released_count
```

```python


unidentified_count=0
# Check if released players are present in the DataFrame
for team, players in released_players.items():
    #print(f"{team} Released Players:")
    for player in players:
        if player not in combined_df['Player'].values:
            # print(f"{player} is not in the DataFrame.")
            unidentified_count+=1
            
        
            
unidentified_count
```

```python
unidentified_count=0

# Add a new column "Released" to the DataFrame
combined_df['Released'] = 'No'

# Mark the players as "Yes" if they are present in the released players list
for club, players in released_players.items():
    for player in players:
        player_found = False
        
        # Try different combinations of names to find a match
        possible_names = [
            player,
            player.split()[0],  # First name only
        ]
        
        if len(player.split()) > 2:
            possible_names.append(player.split()[0] + ' ' + player.split()[1])  # First name and second name
            
            
        
        # Iterate over possible names and check if any matches the DataFrame
        for name in possible_names:
            if name in combined_df['Player'].values:
                combined_df.loc[(combined_df['Player'] == name) & (combined_df['Squad'] == club), 'Released'] = 'Yes'
                player_found = True
                break
        
        if not player_found:
            print(player)
            unidentified_count+=1

unidentified_count
```

```python
released_df = combined_df[combined_df['Released']=='Yes']
released_df.shape
```
